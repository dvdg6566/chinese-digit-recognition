```python
import os, shutil
import pandas as pd
import subprocess
import random
```

```python
os.listdir("./indiv/")
```

```python
data_path = './indiv'
train_path = data_path+"/train"
test_path = data_path+"/test"

val_path = data_path+"/val"
```

```python
def generate_image_path():
    prefix = random.randint(10,31)
    suffix = random.randint(0,14)
    path = "{}-{}.png".format(prefix,suffix)
    return path
    
        

    
for num in os.listdir(train_path):
    if num == ".DS_Store": continue
    for i in range(10):
        cat_path = train_path+"/"+str(num)
        path = generate_image_path()
        print(cat_path)
        print(os.listdir(cat_path))
        while path not in os.listdir(cat_path):
            path = generate_image_path()
        subprocess.call(["mv",cat_path+"/"+path, test_path+"/"+str(num) ])
        
            
```

```python
for num in os.listdir(train_path):
    if num == ".DS_Store": continue
    for i in range(8):
        cat_path = train_path+"/"+str(num)
        path = generate_image_path()
        print(cat_path)
        print(os.listdir(cat_path))
        while path not in os.listdir(cat_path):
            path = generate_image_path()
        subprocess.call(["mv",cat_path+"/"+path, val_path+"/"+str(num) ])
        
```
